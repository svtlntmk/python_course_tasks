````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})


df.memory_usage(index=False).sum()
````

````python
def to_categorical(df, columns):
    for column in columns:
        df[column] = df[column].astype('category')
    return df


columns_to_convert = ['product_name', 'product_category']
df_new = to_categorical(df, columns_to_convert)

df_new.memory_usage(index=False).sum()

#В скільки разів менше памʼяті займає новий датафрейм?
#приблизно в 1.778
````

````python
def calculate_variance(number_list):
    # Calculate the sum of all numbers in the list
    sum_of_numbers = sum(number_list)
    
    # Calculate the mean (average) value of the numbers
    mean_value = sum_of_numbers / len(number_list)
    
    # Calculate the sum of squared differences from the mean
    sum_squared_diff = sum((x - mean_value) ** 2 for x in number_list)
    
    # Calculate the variance
    variance = sum_squared_diff / len(number_list)
    
    return variance
````
